Now let's use a Counter (from collections) to get the counts we want. This is quite a useful utility. Like defaultdict, Counter is just a dictionary (hashtable over key/value pairs) with an additional set of useful methods. In this case, it gives us counts. 

```python
freqs = Counter(words)
```

What's the total of those values?  The size in tokens of the corpus.

```python
N = sum(freqs.values())
```

```python
N
```

Recall that the probability of a word is just its count divided by the total count of tokens.

```python
freqs["dinner"]/N
```

```python
def uniProb(word): return freqs[word]/sum(freqs.values())
```

For a unigram model, the probability of a sentence is the product of the probability of each of the words. Let's start with a utility function. 

```python
def product(nums):
    prod = 1
    for num in nums: prod = prod * num
    return prod
```

```python
def sentenceProb(words):
    return product(uniProb(word) for word in words)
        
```
