```python
import random
```

```python
import bisect 
```

"one fish two fish red fish blue fish black fish blue fish"

```python
words = ["fish", "blue", "one", "two", "red","black"]
```

```python
wordprobs = [6/12, 2/12, 1/12, 1/12, 1/12, 1/12]
```

```python
def getSampleBins(wordprobs):
    index = 0; bins = []
    for w in wordprobs:
        index += w
        bins.append(index)
    return bins
```

```python
def sampleWords(bins):
    rndIndex = random.random()
    return bisect.bisect(bins, rndIndex)
```

```python
getSampleBins(wordprobs)
```

```python
bins = getSampleBins(wordprobs)
```

```python
sampleWords(bins)
```

```python
[[words[sampleWords(bins)] for i in range(12)] for j in range(100)]
```
